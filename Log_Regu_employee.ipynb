```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,f1_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("/content/drive/MyDrive/modified_employee_turnover (1).csv")
data.info()
```

```python
data.head(),data.isnull().sum()
```

```python
X=data.drop(columns='Employee_Turnover')
y=data['Employee_Turnover']
X.shape,y.shape
data.Employee_Turnover.value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=25)
y_train.mean(),y_test.mean()
```

```python
Logistic=LogisticRegression(penalty=None,
                            max_iter=10000,
                            n_jobs=-1)
Logistic.fit(X_train,y_train)
```

```python
Logistic.coef_
```

```python
y_train_pred=Logistic.predict(X_train)
f1=f1_score(y_train,y_train_pred)
print(f1)
```

```python
y_pred=Logistic.predict(X_test)
f1=f1_score(y_test,y_pred)
print(f1)
```

With Regularization

```python
C=np.linspace(0.001,10,20)
```

```python
logistic_l1= LogisticRegressionCV(
    Cs=C,
    penalty='l1',
    solver='liblinear',
    max_iter=10000,
    cv=5,
    scoring='f1',
    n_jobs=-1
)
logistic_l1.fit(X_train,y_train)
```

```python
logistic_l1.Cs
```

```python
best_c=logistic_l1.C_
print(best_c)
```

```python
y_train_pred=logistic_l1.predict(X_train)
f1=f1_score(y_train,y_train_pred)
print("f1 score after l1 regularization:",f1)
```

```python
y_pred=logistic_l1.predict(X_test)
f1=f1_score(y_test,y_pred)
print("f1 score after l2 regularization",f1)
```

```python
coef_l1=pd.DataFrame({
    'Features':X_train.columns,
    'After L1':logistic_l1.coef_.flatten()
})
print(coef_l1.sort_values(by='After L1'))
```

```python
zero_mask=(logistic_l1.coef_==0).flatten()
zero_features=X_train.columns[zero_mask]
zero_features.tolist()
```
